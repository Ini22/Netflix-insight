```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df
```

```python
# Start coding here! Use as many cells as you like
# Subset dataframe to keep movies column only
subset = netflix_df[netflix_df['type'] == 'Movie']

# Filter the years between 1990 and 1999
subset_yr = subset[(subset['release_year'] >= 1990) & (subset['release_year'] <= 1999)]
subset_yr
```

```python
#Visualize data to view distribution of duration of movies released in the 190s
plt.hist(subset_yr['duration'], bins=100)
plt.title("Movie Duration Frequency in the 1990s")
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.grid()
subset_yr['duration'].value_counts()
# duration = 94
```

The most frequent movie duration is 4 minutes

```python
# Subset for the action genre
genre = subset_yr[subset_yr['genre'] == 'Action']

# Count number of movies with duration less than 90
short_movie_count = 0

for duration in genre['duration']:
    if duration < 90:
        short_movie_count += 1
short_movie_count
```

From the analysis the number of action movies which are less than 90 minutes are 7 i.e not a lot. Therefore, most action movies are long.
